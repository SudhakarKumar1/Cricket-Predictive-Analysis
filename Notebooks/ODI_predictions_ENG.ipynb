```python
#Filter dataset for england and australia
dataset_filtered = data[ (data['Team 1']=='Australia') & (data['Team 2']=='England')]
dataset_filtered
```

```python
dataset_filtered.rename(columns = {'Team 1': 'Team_1', 'Team 2' : 'Team_2'}, inplace=True)
```

```python
# Converting categorical values to encoded values
x = dataset_filtered.drop(['Winner','Unnamed: 0','Scorecard','Match Date','Host_Country'],axis =1)
x
```

```python
y = dataset_filtered['Winner']
y
```

```python
x['Team_1']= x['Team_1'].replace({'Australia':0})
x['Team_2']= x['Team_2'].replace({'England':1})
x_filtered = pd.get_dummies(x)
y_filtered = pd.get_dummies(y,drop_first = True)
```

```python
print(x_filtered)
```

```python
# Splitting data into training and test set
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x_filtered,y_filtered,train_size = 0.55,random_state = 0)
```

```python
display(x_train.head())
display(y_train.head())
```

```python
# Model building
from xgboost import XGBClassifier

classifier = XGBClassifier(learning_rate=0.25,n_estimators=500,objective='binary:logistic')

classifier.fit(x_train,y_train)
y_pred = classifier.predict(x_test)
```

```python
# Model accuracy
from sklearn.metrics import roc_auc_score
try:
    display(roc_auc_score(y_test,y_pred))
except ValueError:
    pass
display(x_filtered.head())
```

```python
# Filtering data for current venue

x_current = x_filtered[(x_filtered['Team_2']==1) & (x_filtered['Venue_Team1_Home']==1) &
         (x_filtered['Ground_Melbourne']==1)]

display('Melbourne',classifier.predict(x_current).mean())

x_current = x_filtered[(x_filtered['Team_2']==1) & (x_filtered['Venue_Team1_Home']==1) &
         (x_filtered["Ground_Sydney"]==1)]

display('Sydney',classifier.predict(x_current).mean())

x_current = x_filtered[(x_filtered['Team_2']==1) & (x_filtered['Venue_Team1_Home']==1) &
         (x_filtered['Ground_Perth']==1)]

display('Perth',classifier.predict(x_current).mean())
```
